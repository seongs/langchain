```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Milvus
from langchain.schema import Document
```

```python
docs = [
    Document(
        page_content="많은 과학자들이 공룡을 되살려내고 대혼란을 일으킵니다",
        metadata={"id": 1},
    ),
    Document(
        page_content="레오 디카프리오는 꿈속에서 꿈속에서 길을 잃습니다...",
        metadata={"id": 2},
    ),
    Document(
        page_content="심리학자/탐정이 꿈속에서 꿈속에서 일련의 꿈속에서 길을 잃고 인셉은 이 아이디어를 재사용합니다",
        metadata={"id": 3},
    ),
    Document(
        page_content="보통 크기의 여자들은 아주 건강하고 어떤 남자들은 그들을 따라 합니다",
        metadata={"id": 4},
    ),
    Document(
        page_content="장난감은 살아나고 그렇게 하면서 즐거운 시간을 보냅니다",
        metadata={"id": 5},
    ),
    Document(
        page_content="세명의 남자가 존 안으로 들어가고, 세명의 남자가 존 밖으로 나갑니다",
        metadata={"id": 6},
    ),
]


model_name = "jhgan/ko-sbert-nli"
encode_kwargs = {'normalize_embeddings': True}
hf = HuggingFaceEmbeddings(
    model_name=model_name,
    encode_kwargs=encode_kwargs
)



```

```python
db = Milvus.from_documents(
    docs,
    hf,
    connection_args={"host": "localhost", "port": "19530"},
    collection_name="Movie_test1",
)
```

```python
keyword = "장난감"
result = db.search(query=keyword,search_type="similarity",k=2)
```

```python
result
```

```python
expr = "id in [6]"
pks = db.get_pks(expr)
result = db.delete(pks)
```

```python
new_docs = [
    Document(page_content="로봇 장난감", metadata={"id": 7}),
    Document(page_content="감자", metadata={"id": 8}),
    Document(page_content="파도", metadata={"id": 9}),
]
upserted_pks = db.upsert(pks, new_docs)
```

```python
search_params = {"metric_type": "COSINE"}
```
